```python
# Scatter Plot: Genre and imdb rating
plt.figure(figsize=(17, 10))
sns.scatterplot(data=df, x="Genre", y="IMDB_Rating", hue="Meta_score", legend=False)
plt.title("Scatter Plot: the rating of based upon the Genre ")
plt.xlabel("Genre")
plt.ylabel("IMDB_Rating")
plt.show()
```

```python

numeric_columns = df[["Released_Year", "Runtime", "IMDB_Rating", "Meta_score", "No_of_Votes", "Gross"]].select_dtypes(include=[float, int])
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="IMDB_Rating", color="blue")
plt.title("Histogram: Distribution of IMDB Rating  Index")
plt.xlabel("IMDB_Rating Index")
plt.ylabel("no of movies")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y="Genre", color="purple")
plt.title("Box Plot:Movies popular based on rating ")
plt.ylabel("Genre")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Genre", y="IMDB_Rating")
plt.title("Bar Plot: The Rating and Genre")
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df.nsmallest(10, "IMDB_Rating")
plt.figure(figsize=(10, 6))
plt.plot(data["Genre"], data["IMDB_Rating"], color="green")
plt.title("Line Graph: IMDB Rating vs Genre")
plt.xlabel("Genre")
plt.ylabel("IMDB Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df.nlargest(5, "IMDB_Rating")
plt.figure(figsize=(8, 8))
plt.pie(data["IMDB_Rating"], labels=data["Genre"], colors=sns.color_palette("pastel"))
plt.title("Pie Chart:The imdb rating based upon the movie genre")
plt.show()
```
